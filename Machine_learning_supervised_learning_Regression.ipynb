# regression lineaire simple

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression # pour faire du lachine learning
```

```python
house_size = np.array([50, 75, 100, 125, 150]).reshape(-1,1) #taille des maisons en m2
house_price = np.array([150000, 200000, 250000, 300000, 350000]) #pris des maisons en $
```

```python
# Initialiser le modèle de régression linéaire
model = LinearRegression()
# Entraîne le modèle à l'aide de l'ensemble de données
model.fit(house_size, house_price)
```

```python
#predire le prix des maisons en fonction de la taille des maisons
predicted_price = model.predict(house_size)

#.predict(X) → Utilise l'équation apprise pour prédire Y (prix des maisons) pour X (taille des maisons).
```

```python
#calcul de la difference pour voir les erreurs
diff = house_price - predicted_price

#imprimer les diffs
print('diff between actual and predicted price :',diff)
```

# regression lineaire multiple

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

Étape 2 : Préparation des données
Nous créons un ensemble de données avec trois variables d'entrée (taille de la maison, nombre de chambres et âge de la maison) et une variable cible (prix de la maison).

```python
# Définir les variables indépendantes (taille du logement, nombre de chambres, âge du logement)
house_features = np.array([
    [50, 2, 10],
    [75, 3, 5],
    [100, 3, 8],
    [125, 4, 4],
    [150, 4, 2]
])
# Définir la variable dépendante (prix des maisons en $)
house_prices = np.array([150000, 200000, 250000, 300000, 350000])
```

**Étape 3 : Entraînement du modèle de régression linéaire multiple**

Maintenant, nous créons et entraînons le modèle à l'aide de notre ensemble de données

```python
# initialiser le model de regression lineaire
model = LinearRegression()

# entrainer le model a l'aide de l'ensemble de données
model.fit(house_features, house_prices)
```

**Étape 4 : Récupération de l'équation apprise**

Après l'apprentissage, le modèle trouve les meilleurs coefficients (valeurs m) et l'interception (b).

```python
# Obtenir les coefficients et l'interception
m_values = model.coef_  # Pentes
b = model.intercept_  # Intercept
# Afficher l'équation
print(f"Equation de la droite de régression : Y = {m_values[0] :.2f}X1 + {m_values[1] :.2f}X2 + {m_values[2] :.2f}X3 + {b :.2f}")
```

**Etape 5 : Faire des prédictions**

Maintenant que le modèle est entraîné, nous pouvons prédire les prix des maisons sur la base de nouvelles données.

```python
# Nouvelles données de maison pour la prédiction (Taille, Chambres, Âge)
new_house = np.array([[120, 3, 6]])
# Prédire le prix
predicted_price = model.predict(new_house)
# Afficher la prédiction
print(f"Predicted House Price: ${predicted_price[0]:.2f}")
```

**Étape 6 : Visualisation de la relation**

Puisque nous avons trois caractéristiques, nous pouvons visualiser l'effet d'une caractéristique à la fois tout en gardant les autres constantes.

```python
plt.scatter(house_features[:, 0], house_prices, color='blue', label='Actual Prices')  # Actual prices
plt.scatter(new_house[:, 0], predicted_price, color='red', label='Predicted Price', marker='x', s=100)  # Predicted price
plt.xlabel("House Size (m²)")
plt.ylabel("House Price ($)")
plt.title("Effect of House Size on Price")
plt.legend()
plt.show()
```

# évaluation des performances

**Mean Absolute Error (MAE)**

```python
from sklearn.metrics import mean_absolute_error
# Exemple de données
Y_actual = [100, 200, 300, 400, 500]
Y_predicted = [110, 190, 290, 410, 490]
# Calcul du MAE
mae = mean_absolute_error(Y_actual, Y_predicted)
print("Mean Absolute Error (MAE) :", mae)
```

**Mean Squared Error (MSE)**

```python
from sklearn.metrics import mean_squared_error

Y_actual = [100, 200, 300, 400, 500]
Y_predicted = [110, 190, 290, 410, 490]
# Calcul de l'erreur quadratique moyenne
mse = mean_squared_error(Y_actual, Y_predicted)
print("Mean Squared Error (MSE) :", mse)
```

**Root Mean Squared Error (RMSE)**

```python
import numpy as np

# Calculer la RMSE
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE) :", rmse)
```

**R-Carré (R²) - Coefficient de détermination**

```python
from sklearn.metrics import r2_score

Y_actual = [100, 200, 300, 400, 500]
Y_predicted = [110, 190, 290, 410, 490]
# Calculer le score R²
r2 = r2_score(Y_actual, Y_predicted)
print("R-Squared (R²) :", r2)
```

#  Implémentation de la régression Ridge

Elle est particulièrement efficace pour remédier au surajustement

1 Importation des bibliothèques requises

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
```

 2.creer et preparer le jeu de données

```python
# Exemple de jeu de données (taille des maisons vs prix)
X = np.array([[500], [600], [700], [800], [900], [1000], [1100], [1200]])
Y = np.array([150000, 180000, 210000, 240000, 270000, 300000, 330000, 360000])
```

3 Diviser les données en ensembles d'entraînement et de test

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

4 entrainement du model de regression Ridge

```python
ridge_model = Ridge(alpha=1.0)  ; # λ = 1.0 (force de régularisation)
ridge_model.fit(X_train, Y_train)
```

5 faire des predictions

```python
Y_pred = ridge_model.predict(X_test)
print("Prix prédits :", Y_pred)
```

6 Visualisation du modèle de régression Ridge

```python
plt.scatter(X, Y, color='blue', label='Actual Prices')
plt.plot(X, ridge_model.predict(X), color='red', lw=2, label='Ridge Regression')
plt.xlabel("Taille de la maison (pieds carrés)")
plt.ylabel("Prix du logement ($)")
plt.title("Prédiction du prix des maisons à l'aide de la régression Ridge")
plt.legend()
plt.show()
```

5 Comparaison entre Ridge et la régression linéaire

```python
from sklearn.linear_model import LinearRegression
# Former un modèle de régression linéaire simple
linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
# Comparer les coefficients
print("Coefficients de régression linéaire :", linear_model.coef_)
print("Coefficients de régression ridge :", ridge_model.coef_)
```

- Observations clés

 - La régression ridge rétrécit les coefficients, ce qui empêche l'ajustement excessif.

 - La régression linéaire ne pénalise pas les grands coefficients, ce qui peut entraîner une instabilité.

# Implémentation d'ElasticNet en Python

1 Importer les bibliothèques nécessaires

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
```

2 Charger et préparer les données

Nous utiliserons un ensemble de données sur le prix des maisons où nous prédisons le prix en fonction de la taille de la maison, du nombre de chambres et du score de localisation.

```python
# Exemple d'ensemble de données (taille de la maison, chambres, score de localisation par rapport au prix)
# Sample dataset (house size, bedrooms, location score vs price)
data = {
    "Size": [1500, 1800, 2100, 2500, 2800, 3100, 3500, 4000],
    "Bedrooms": [3, 3, 4, 4, 5, 5, 6, 6],
    "LocationScore": [7, 8, 6, 9, 7, 8, 6, 9],
    "Price": [300000, 350000, 400000, 450000, 500000, 550000, 600000, 700000]
}

df = pd.DataFrame(data)

# Diviser l'ensemble de données en caractéristiques (X) et en variables cibles (Y)
X = df.drop(columns=["Price"])
Y = df["Price"]

# Diviser le jeu de données en ensembles d'entraînement (80 %) et de test (20 %)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

 3. Entraîner un modèle ElasticNet



```python
# Définir un modèle ElasticNet avec alpha et l1_ratio (paramètre de mélange)
model = ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=42)

# Entraînement du modèle
model.fit(X_train, Y_train)

# Prédire sur l'ensemble de test
Y_pred = model.predict(X_test)

# Calculer le root mean squared error (RMSE)
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
print("Root Mean Squared Error (RMSE) :", rmse)
```

6.Réglage des hyperparamètres pour ElasticNet

- Hyperparamètres clés dans ElasticNet
 - alpha (λ) - Contrôle la force de régularisation globale.
 - l1_ratio (α) - Équilibre les pénalités Lasso (L1) & Ridge (L2) (0 = Ridge, 1 = Lasso).
 - max_iter - Itérations maximales pour l'optimisation.

# regression des series temporelles

- ARIMA(AutoRegressive Integrated Moving Average)
- implementation de l'ARIMA


```python
#importer les bibliotheques necessaire

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
#charger et preparer les données de serie temporelle

# Échantillonner des données de ventes mensuelles
data = {
    "Month": pd.date_range(start="2023-01-01", periods=12, freq="M"),
    "Sales": [5000, 5200, 5300, 5500, 5800, 6000, 6200, 6400, 6600, 7000, 7500, 8000]
}
df = pd.DataFrame(data)
# Définir la date comme index
# carLes modèles de séries temporelles
# requièrent un indice de date pour comprendre l'ordre des observations.

df.set_index("Month", inplace=True)
```

```python
# former un modèle ARIMA

# Définir un modèle ARIMA (p, d, q)
model = ARIMA(df, order=(1, 1, 1))  # (AutoRegressive=1, Differencing=1, Moving Average=1)

# Ajuster le modèle
arima_result = model.fit()

# Imprimer le résumé du modèle
print(arima_result.summary())

#p (terme autorégressif) → Utilise les valeurs passées.
#d (terme de différenciation) → Supprime les tendances.
#q (terme de moyenne mobile) → Utilise les erreurs passées pour améliorer les prédictions.
```

```python
# fire des predictions

# Prévision des 3 prochains mois
forecast = arima_result.forecast(steps=3)
print("Ventes prévues :", forecast)
```

```python
#Évaluer les performances du modèle


# Diviser les données en formation (9 premiers mois) et en test (3 derniers mois)
train_size = 9
train, test = df[:train_size], df[train_size :]

# Ajuster ARIMA sur les données d'entraînement
model_train = ARIMA(train, order=(1, 1, 1)).fit()

# Prédire sur l'ensemble de test
predictions = model_train.forecast(steps=len(test))

# Calculer les erreurs
mae = mean_absolute_error(test, predictions)
mse = mean_squared_error(test, predictions)
rmse = np.sqrt(mse)

print("Erreur absolue moyenne (MAE) :", mae)
print("Erreur quadratique moyenne (EQM) :", mse)
print("Erreur quadratique moyenne (RMSE) :", rmse)
```

```python
#Visualisation des prédictions
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["Sales"], label="Actual Sales", marker="o")
plt.plot(test.index, predictions, label="Predicted Sales", linestyle="dashed", marker="x")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.title("ARIMA Time Series Regression: Sales Forecast")
plt.legend()
plt.show()
```
